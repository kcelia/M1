## Programmer un joueur d'Awélé

### Structure de données

Pour représenter une position d'Awélé en Python. Une *position* est définie par les éléments suivants :
- une *taille*, en nombre de colonnes (dans l'Awélé original, il y a 6 colonnes). On note $n$ ce nombre de colonnes dans la suite de ce texte ;
- un *tablier* qui correspond au plateau de jeu et donne le nombre de graines dans chaque case ;
- le *trait* qui correspond au camp devant jouer dans la position donnée. Il y a
  deux camps: "NORD et "SUD" ;
- les nombres de *graines* déjà ramassées par chacun des joueurs.


En Python, une telle position est représentée en tant que dictionnaire qui associe à chaque nom de champ sa valeur.

Dans ce qui suit, on distingue :
- un coup *jouable* qui est un numéro de colonne correct (valeur comprise entre 1
  et $n$), et pour lequel la case correspondante contient au moins une graine ;
- un coup *autorisé* qui est un coup à la fois jouable et permettant d'atteindre
  une position légale du jeu, c'est-à-dire une position qui respecte la règle obligeant à
  laisser au moins une graine dans le camp de son adversaire.

** Représentation du tablier en Python**

Pour représenter le tablier en Python, on utilise une liste d'entiers, chaque entier correspondant au nombre de graines dans une case. 

Soit un tablier de $n$ colonnes.

Pour le camp SUD :
- le nombre de graines de la case de la colonne 1 est stocké en position $0$ dans la liste Python ;
- le nombre de graines de la case de la colonne  2 est stocké en position $1$ dans la liste Python ;
- ...
- le nombre de graines de la case de la colonne  $n$ est stocké en position $n-1$ dans la liste Python.

Pour le camp NORD :
- le nombre de graines de la case de la colonne  1 est stocké en position $2n-1$ dans la liste Python ;
- le nombre de graines de la case de la colonne  2 est stocké en position $2n-2$ dans la liste Python ;
- ...
- le nombre de graines de la case de la colonne  $n$ est stocké en position $n$ dans la liste Python.

(il est conseillé de faire un schéma pour bien comprendre).

Cette représentation à l'avantage de favoriser le parcours des cases du tablier : en parcourant la liste par positions croissantes, on retrouve le parcours en sens inverse des aiguilles d'une montre. Ce parcours peut alors se faire en Python en incrémentant de 1 la position pour passer à la case suivante, et en réalisant un calcul modulo $2n$.

### Fonctions fournies

Plusieurs fonctions sont fournies dans ce qui suit :
- <tt>initialise(n)</tt> : pour définir la position initiale de l'Awélé. L'argument est un entier <tt>n</tt> qui donne le nombre de colonnes du tablier du jeu ;
- <tt>affichePosition(position)</tt> : affichage de façon textuelle d'une position d'Awélé ;
- <tt>clonePosition(position)</tt> : recopie d'une position (clonage) pour pouvoir y
  apporter des modifications sans altérer la position originale ;
- <tt>joueCoup(position,coup)</tt> : renvoie la position obtenue une fois le coup joué
  dans la position donné. Cette fonction fait l'hypothèse que le coup donné est un coup
  jouable et elle ne vérifie pas que la position résultante est bien une position légale
  du jeu.


```python
# - - - - - - - - - - - - - - - TYPES UTILISES
# POSITION : dictionnaire non vide qui contient différentes informations sur
#            une position d'Awélé, associées au nom de leur champ.
# COUP : valeur entière comprise entre 1 et le nombre de colonnes du tablier
#
#
# - - - - - - - - - - - - - - - INITIALISATION
def initialise(n):
    """ int -> POSITION
        Hypothèse : n > 0
        initialise la position de départ de l'awélé avec n colonnes avec 4 graines dans chaque case.
    """
    position = dict()                                 # initialisation
    position['tablier'] = [4 for k in range(0, 2*n)]  # on met 4 graines dans chaque case
    position['taille'] = n                            # le nombre de colonnes du tablier
    position['trait'] = 'SUD'                         # le joueur qui doit jouer: 'SUD' ou 'NORD'
    position['graines'] = {'SUD':0, 'NORD':0}         # graines prises par chaque joueur
    return position

# - - - - - - - - - - - - - - - AFFICHAGE (TEXTE)
def affichePosition(position):
    """ POSITION ->
        affiche la position de façon textuelle
    """
    print('* * * * * * * * * * * * * * * * * * * *')
    n = position['taille']
    buffer = 'col:'
    for i in range(0,n):
        buffer += ' ' + str(i+1) + ' \t'
    print(buffer)
    print('\t\tNORD (prises: '+str(position['graines']['NORD'])+')')
    print('< - - - - - - - - - - - - - - -')
    buffer = ''
    for i in range(2*n-1,n-1,-1):   # indices n..(2n-1) pour les cases NORD
        buffer += '\t[' + str(position['tablier'][i]) + ']'
    print(buffer)
    buffer = ''
    for i in range(0,n):            # indices 0..(n-1) pour les cases SUD
        buffer += '\t[' + str(position['tablier'][i]) + ']'
    print(buffer)
    print('- - - - - - - - - - - - - - - >')
    print('\t\tSUD (prises: '+str(position['graines']['SUD'])+')')
    print('-> camp au trait: '+position['trait']);

# - - - - - - - - - - - - - - - CLONAGE
import copy
def clonePosition(position):
    """ POSITION -> POSITION
        retourne un clone de la position
        (qui peut être alors modifié sans altérer l'original donc).
    """
    leclone = dict()
    leclone['tablier'] = copy.deepcopy(position['tablier'])
    leclone['taille']  = position['taille']
    leclone['trait']   = position['trait']
    leclone['graines'] =  copy.deepcopy(position['graines'])
    return leclone

# - - - - - - - - - - - - - - - JOUE UN COUP
def joueCoup(position,coup):
    """ POSITION * COUP -> POSITION
        Hypothèse: coup est jouable.

        Cette fonction retourne la position obtenue une fois le coup joué.
    """
    nouvelle_pos = clonePosition(position)   # on duplique pour ne pas modifier l'original
    n = nouvelle_pos['taille']
    trait = nouvelle_pos['trait']
    # on transforme coup en indice
    if trait == 'SUD':
        indice_depart = coup-1
    else:
        indice_depart = 2*n-coup
    # retrait des graines de la case de départ
    nbGraines = nouvelle_pos['tablier'][indice_depart]
    nouvelle_pos['tablier'][indice_depart] = 0
    # on sème les graines dans les cases à partir de celle de départ
    indice_courant = indice_depart
    while nbGraines > 0:
        indice_courant = (indice_courant + 1) % (2*n)
        if (indice_courant != indice_depart):              # si ce n'est pas la case de départ
            nouvelle_pos['tablier'][indice_courant] += 1   # on sème une graine
            nbGraines -= 1
    # la case d'arrivée est dans le camp ennemi ?
    if (trait == 'NORD'):
        estChezEnnemi = (indice_courant < n)
    else:
        estChezEnnemi = (indice_courant >= n)
    # réalisation des prises éventuelles
    while estChezEnnemi and (nouvelle_pos['tablier'][indice_courant] in range(2,4)):
        nouvelle_pos['graines'][trait] += nouvelle_pos['tablier'][indice_courant]
        nouvelle_pos['tablier'][indice_courant] = 0
        indice_courant = (indice_courant - 1) % (2*n)
        if (trait == 'NORD'):
            estChezEnnemi = (indice_courant < n)
        else:
            estChezEnnemi = (indice_courant >= n)
    # mise à jour du camp au trait
    if trait == 'SUD':
        nouvelle_pos['trait'] = 'NORD'
    else:
        nouvelle_pos['trait'] = 'SUD'
    return nouvelle_pos
# - - - - - - - - - -
```

### Fonctions simples

<font color="RED" size="+1">**[Q]**</font> Ecrire les fonctions suivantes :
- <tt>coupJouable(position,nombre)</tt>: rend le booléen <tt>True</tt> si le nombre donné est
  jouable dans la position donnée, c'est-à-dire si ce nombre est un entier compris entre 1
  et <tt>n</tt> et si de plus la case correspondante du tablier (pour le camp qui doit donc jouer
  dans la position) contient au moins une graine.
- <tt>coupAutorise(position,coup)</tt>: rend la position obtenue en jouant <tt>coup</tt>
  dans la position donnée. Cette fonction retourne le booléen <tt>False</tt> si le coup
  n'est pas jouable ou si la position résultante n'est pas une position légale du jeu.
- <tt>positionTerminale(position)</tt>: rend le booléen <tt>True</tt> si la position est
  terminale, c'est-à-dire si aucun coup n'est jouable, ou si tout coup jouable ne permet
  pas d'atteindre une position légale ou bien si l'un des deux joueurs a gagné au moins 25
  graines.

```python
def coupJouable(position, nombre):
    """POSITION * int -> bool
        retourne True si le nombre donné est jouable dans la position donnée
    """
    # On vérifie que le nombre correspond à un coup
    if 0 < nombre <= position["taille"]:
        # On vérifie la présence de graines dans la case
        if position["trait"] == 'SUD':
            return (position["tablier"][nombre-1] > 0)
        else:
            return (position["tablier"][2*position["taille"] - nombre] > 0)
    return False
                                                                                                                                                
    
def coupAutorise(position, coup):
    """POSITION * COUP -> POSITION
        retourne la position obtenue en jouant coup dans la position donnée.
        Si le coup n'est pas jouable ou non légale, retourne False.
    """
    
    if coupJouable(position, coup):
        new_pos = joueCoup(position, coup)
        # new_pos etant la position du tour suivant, c'est le tour de l'adversaire
        # d'où si le trait de new_pos est SUD alors on verifie qu'il y a au moins 
        # une graine de son côté
        if new_pos["trait"] == 'SUD':
            if (sum(new_pos["tablier"][:new_pos["taille"]]) > 0):
                return new_pos
        else:
            if (sum(new_pos["tablier"][new_pos["taille"]:]) > 0):
                return new_pos
    return False


def positionTerminale(position):
    """POSITION -> bool
        retourne True si la position est terminale
    """
    # Nombre de graines supérieur à 25 dans l'un des camps
    if (position['graines']['NORD'] >= 25 or position['graines']['SUD'] >= 25) : 
        return True 
    
    # On vérifie si le joueur peut jouer un coup (pas besoin de vérifier le camp)
    # Cela se fait automatiquement par coup autorisé
    for i in range(1,position["taille"]+1):
        res = coupAutorise(position,i) 
        if (res != False):
                return False
    return True   
    
```

  possibles dans la position donnée. Cette fonction rend la valeur 0 si aucun coup
  autorisé n'est possible.

```python
from random import choice
def choixAleatoire(position):
    """POSITION -> COUP
        retourne un coup autorisé pour la position donnée de manière aléatoire
    """
    
    if position["trait"] == 'SUD':
        cases = position["tablier"][:position["taille"]]
    else:
        cases = position["tablier"][position["taille"]:]
    # Liste de coups autorisés
    coups = [i+1 for i, case in enumerate(cases) if coupAutorise(position, i+1)]
    return choice(coups)
```

## Joueur artificiel simple

On considère la fonction d'évaluation suivante de la position $P$ :
$$f(P) = \left.
  \begin{array}{ll}
        +1000      & \mbox{si } P \mbox{ est gagnée par sud} \\
        -1000      & \mbox{si } P \mbox{ est gagnée par nord} \\
        (2f_g(P,sud) + f_{12}(P,nord)) - (2f_g(P,nord) + f_{12}(P,sud))& \mbox{sinon }
   \end{array} \right.$$

avec
- $f_g(P,c)$ : nombre de graines déjà gagnées par le camp $c$ dans la position $P$
- $f_{12}(P,c)$ : nombre de cases du camp $c$ contenant 1 ou 2 graines, dans la position $P$.



<font color="RED" size="+1">**[Q]**</font> Ecrire la fonction <tt>evaluation(position)</tt> qui rend la valeur de l'évaluation de la position obtenue en utilisant la fonction $f$.

```python
def evaluation(position):
    """POSITION -> int
        retourne la valeur de l'évaluation de la position obtenue en utilisant la fonction f
    """
    
    if position["graines"]["SUD"] >= 25:
        return 1000
    elif position["graines"]["NORD"] >= 25:
        return -1000
    else:
        # On calcule le nombre de cases du camp contenant 1 ou 2 graines
        # PS : On ne vérifie pas la contiguité de ces cases
        nb_sud = sum(case in range(1,3) for case in position["tablier"][:n])
        nb_nord = sum(case in range(1,3) for case in position["tablier"][:n])
        return 2 * position["graines"]["SUD"] + nb_nord - 2 * position["graines"]["NORD"] - nb_sud
```

```python
from tkinter import *


class Interface(Frame):
    """ représente la fenêtre principale
        widgets stockés en tant qu'attributs
    """
    
    def __init__(self, fenetre, **kwargs):
        Frame.__init__(self, fenetre, width=800, height=500, padx=50, pady=50, **kwargs)
        self.pack(fill=BOTH)
        
        #### MENU PRINCIPAL - INITIALISATION BOUTONS
        self.bouton_jouer = Button(self, text="J vs CPU", width=50, height=3, font=2)
        self.bouton_jouer2 = Button(self, text="J vs J", width=50, height=3, font=2)
        self.bouton_regles = Button(self, text="Règles du jeu", width=50, height=3, font=2)
        self.bouton_quitter = Button(self, text="Quitter", width=50, height=3, font=2, command=fenetre.quit)
        self.label = Label(self, text="IAMSI 2017-2018 - Awélé")
        
        #### OPTIONS DE DIFFICULTES
        self.facile = Button(self, text="Débutant", width=50, height=3, font=2)
        self.moyen = Button(self, text="Amateur", width=50, height=3, font=2)
        self.difficile = Button(self, text="Maître", width=50, height=3, font=2)
        
        ### REGLES DU JEU
        self.frame_regles = LabelFrame(self, text="Regles du jeu", padx=20, pady=20, font=1)
        self.regles = Label(self.frame_regles, text=REGLES, padx=30, wraplength=400, font=1).pack()
        self.bouton_regles.configure(command=self.afficher_regles)
        
        #### BOUTON DE RETOUR
        self.bouton_retour = Button(self, text="Retour", width=20, height=3)
        self.bouton_retour.configure(command=self.menu)
        
        ### JOUER ET DIFFICULTES
        self.bouton_jouer.configure(command=self.afficher_difficultes)
        self.bouton_jouer2.configure(command=self.initialise_partie)
        self.facile.configure(command=lambda: self.lancer_partieJvCPU(1))
        self.moyen.configure(command=lambda: self.lancer_partieJvCPU(2))
        self.difficile.configure(command=lambda: self.lancer_partieJvCPU(3))
        
        #### AFFICHAGE
        self.menu()
        
        
    def menu(self):
        """ Affichage du menu principal"""
        self.clear()
        l = Label(self, text="AWALÉ", font=("Tahoma", 32, "bold"), pady=20)
        l.pack()
        self.bouton_jouer.pack()
        self.bouton_jouer2.pack()
        self.bouton_regles.pack()
        self.bouton_quitter.pack()
        self.label.pack()
        
        ### CPU ne joue pas pour le moment
        self.CPU = False
    
    
    def afficher_difficultes(self):
        """ Affichage des difficultés """
        self.clear()
        self.facile.pack()
        self.moyen.pack()
        self.difficile.pack()
        self.bouton_retour.pack()
        
        
    def afficher_regles(self):
        """ Affichage des règles """
        self.clear()
        self.frame_regles.pack()
        self.bouton_retour.pack()
    
    
    def refresh_position(self, position, coup):
        """ POSITION * COUP -> bool
            Mise a jour de l'affichage de la position 
            retourn False si la position est terminale
        """
        
        #### On retire le label affamé si présent
        self.affame.pack_forget()
        
        #### Verification coup autorisé
        new_pos = coupAutorise(position, coup)
        if new_pos == False:
            self.affame.pack()
        self.position = new_pos
        
        #### Mise a jour des labels
        self.score["text"] = "SUD : %s ------ %s : NORD" % \
                    (self.position["graines"]["SUD"],self.position["graines"]["NORD"])
        self.tour["text"] = "A %s de jouer !" % (self.position["trait"])
        
        #### Verification position terminale
        if positionTerminale(self.position):
            Label(self, text="GAME OVER", font=("Impact",32), pady=20).pack()
            self.lock_board
            return False
            
        
        #### Mise à jour des cases (boutons)
        for i in range(self.position["taille"]):
            self.nord[i]["text"] = self.position["tablier"][2*self.position["taille"]-i-1]
        for i in range(self.position["taille"]):
            self.sud[i]["text"] = self.position["tablier"][i]
        
        #### Mise à jour de l'accès aux cases
        if self.position["trait"] == 'SUD':
            for case in self.nord:
                case["state"] = "disabled"
            for i, case in enumerate(self.sud):
                case["state"] = "normal" if self.position["tablier"][i] != 0 else "disabled"
        else:
            for case in self.sud:
                case["state"] = "disabled"
            for i, case in enumerate(self.nord):
                case["state"] = "normal" if self.position["tablier"][2*self.position["taille"]-i-1] != 0 else "disabled"
        
        if self.CPU == True:
            self.jeuCPU()
        
        return True
        
        
    def initialise_partie(self):
        """ Lancement d'une partie 
            Hypothèse : SUD commence toujours
        """
        #### Nettoyage et création d'un frame pour le plateau
        self.clear()
        frame1 = Frame(self, padx=40, pady=40)
        #### Position
        self.position = initialise(6)
        #### Initialisation labels
        self.tour = Label(self, text="A %s de jouer !" % (self.position["trait"]), font=5)
        self.score = Label(self, text="SUD : %s ------ %s : NORD" % (self.position["graines"]["SUD"], \
                                       self.position["graines"]["NORD"]), font=("Calibri", 24, "bold"), pady=30)
        #### on prépare le label avertissement pour ne pas affamer l'adversaire
        self.affame = Label(self, text="Stop, vous allez affamer l'adversaire, le pauvre !",\
                            font=("Calibri", 20, "italic"))
        
        
        #### Initialisation des cases (buttons)
        self.sud, self.nord = [], []
        j = 0
        
        while j < self.position["taille"]:
            # POUR NORD
            nb_graines = self.position["tablier"][2*self.position["taille"]-j-1]
            self.nord.append(Button(frame1, text="%d" % (nb_graines), \
                              borderwidth=5, width=15, height=6, bg="white",\
                                    relief=RAISED, state="disabled", font="bold"))
            self.nord[j].grid(row=1,column=j+1)
            # POUR SUD
            nb_graines = self.position["tablier"][j]
            self.sud.append(Button(frame1, text="%d" % (nb_graines),\
                              borderwidth=5, width=15, height=6, bg='yellow', relief=RAISED, font="bold"))
            self.sud[j].grid(row=2,column=j+1)
            
            #self.nord[j].configure(command=lambda:self.refresh_position(self.position, j+1))
            #self.sud[j].configure(command=lambda:self.refresh_position(position, j+1))
            #### n'a pas l'air de fonctionner, prend toujours la derniere valeur de j
            j+=1
            
        #### Solution de secours suite au problème de lambda
        self.nord[0].configure(command=lambda : self.refresh_position(self.position, 1))
        self.nord[1].configure(command=lambda : self.refresh_position(self.position, 2))
        self.nord[2].configure(command=lambda : self.refresh_position(self.position, 3))
        self.nord[3].configure(command=lambda : self.refresh_position(self.position, 4))
        self.nord[4].configure(command=lambda : self.refresh_position(self.position, 5))
        self.nord[5].configure(command=lambda : self.refresh_position(self.position, 6))
        
        self.sud[0].configure(command=lambda : self.refresh_position(self.position, 1))
        self.sud[1].configure(command=lambda : self.refresh_position(self.position, 2))
        self.sud[2].configure(command=lambda : self.refresh_position(self.position, 3))
        self.sud[3].configure(command=lambda : self.refresh_position(self.position, 4))
        self.sud[4].configure(command=lambda : self.refresh_position(self.position, 5))
        self.sud[5].configure(command=lambda : self.refresh_position(self.position, 6))

        #### On pack les widgets
        self.tour.pack()
        frame1.pack()
        self.score.pack()
        self.bouton_retour.pack(side=BOTTOM)
        
        
    def lock_board(self):
        """ Bloque le plateau le temps que le CPU joue """
        for case in self.nord:
                case["state"] = "disabled"
        for case in self.sud:
                case["state"] = "disabled"
        
        
    def lancer_partieJvCPU(self, diff):
        """ Lance une partie entre un joueur et le CPU """
        self.CPU = True # CPU joue 
        self.diff = diff
        self.initialise_partie()
        self.coup_joue = Label(self, font=("Calibri", 16), pady=10)
        self.coup_joue.pack()
    
        
    def jeuCPU(self):
        """ Coup joué par le CPU et qui vérifie si la position est terminale"""
        if self.position["trait"] == 'NORD':
            self.lock_board()
            coup = 0 # Juste pour initialiser
            if self.diff == 1:
                coup = choixAleatoire(self.position)
            elif self.diff == 2:
                coup = choixAlphaBeta(self.position, 1)
            else:
                coup = choixAlphaBeta(self.position, 7)
            
            self.coup_joue["text"] = "NORD a joué le coup %d" % (coup)
            if self.refresh_position(self.position, coup) == False:
                return
        
        
    def clear(self):
        """ Vide la fenêtre de tous les widgets """
        for child in self.children.values():
            # "unpack" les widgets contenus dans l'objet
            child.pack_forget()
            
            
```

```python
def nouvelleRegle(position,coup):
    """
    La fonction nouvelleRegle permet de definir une nouvelle regle du jou qui est :
    un coup qui retire toutes les graines du camp de 
     l'adversaire peut être quand même joué mais dans ce cas, aucune prise n'est effectuée.
     
    """
    #l'algorithme est le suivant 
    #On joue le coup proposé par le joueur X
    #On regarde le camp adverse si toutes les graines ont été ramassé par le joueur X
    #si le tablier du camp adverse ne comporte plus aucune graine, on annule le coup du joueur X, 
    #on distribue les graines que X a jouer sans utiliser la fonction joueCoup definie precedemment
    #le joueur X ne ramasse pas de graine et son score n'est pas modifié
    #on change le trait de la position 
    #sinon (le tablier du camp adverse comporte des graines) on retourne cette nouvelle position 
    
    x = joueCoup(position,coup)
    
    # pour fixer les bords 
    if position["trait"] == "SUD":
        (inf,sup ) = (x["taille"],2*x["taille"])
        coup=coup-1
    else:
        (inf,sup) = (0,x["taille"])
        coup=2*x["taille"]-coup
        
    #teste pour savoir si le camp adverse possede ou pas des graines dans son tablier 
    if sum(x["tablier"][inf : sup ]) == 0:
        #si oui, on distribut les graines
        nbr_graine = position["tablier"][coup]
        position["tablier"][coup] = 0
        i = coup
        while(nbr_graine > 0 ):
            if (i)%( 2*position["taille"]) != coup:
                print((i)%( 2*position["taille"]))
                position["tablier"][(i)%( 2*position["taille"])] += 1 
                nbr_graine -= 1
            i += 1 
        #on donne la main a l autre pour le prochain tour
        position["trait"] = "SUD" if position["trait"]=="NORD" else "NORD"
            
        return position
    else:
        #sinon, ie le tablier du camp possede des graines, donc on retroune le resultat de la fonction joueCoup
        return x

def coupAutoriseNouvelleRegle(position,coup):
    """la fonction est une variante de coupAutorise celle ci permet d'instauré une nouvelle regle 
    """
    if not coupJouable(position,coup) :
        return False 
        
    return nouvelleRegle(position,coup)

```

## Annexe : présentation du jeu d'Awélé

L'Awélé est un jeu d'origine africaine qui se joue sur un plateau (ou tablier) où chacun des 2 joueurs (Sud et Nord) possède 6 cases.  Les règles de ce jeu  sont simples et facilement implémentables.

Voir la page wikipédia (http://fr.wikipedia.org/wiki/Awalé) qui détaille les règles que nous utilisons ici, ainsi que le site http://s.helan.free.fr/awale/lejeu/jouer/ qui permet de s'initier en jouant
  contre un joueur artificiel.

### Règles du jeu
Dans la position de départ, toutes les cases sont remplies avec 4 graines.  Le joueur dont
c'est le tour de jeu (on considère que le camp Sud débute toujours la partie.) choisit une de ses cases contenant des graines, et en retire toutes les graines qu'elle contient.  Il sème alors ces graines une par une dans les cases suivantes dans le sens inverse des aiguilles d'une montre.  Au cours de cette pose, s'il repasse sur la case qui contenait les graines au départ, il n'y dépose pas de graine.

Si la dernière graine est semée dans une case de l'adversaire contenant 2 ou 3 graines après la pose, les graines dans cette case sont capturées par le joueur et elles sont alors retirées du jeu (elles sont mises dans le butin du joueur).  Dans ce cas, si l'avant-dernière case est aussi une case ennemie contenant 2 ou 3 graines, les graines en sont aussi capturées, et ainsi de suite tant que des prises sont possibles (toujours dans le camp ennemi).

Il est à noter qu'un coup n'effectue pas forcément une prise, mais si une prise existe à l'issue d'un coup, elle doit obligatoirement être réalisée totalement.

A l'issue du coup ainsi exécuté, il doit rester au moins une graine dans l'une des cases de l'adversaire (il ne faut pas ''affamer'' l'adversaire) sinon la position est considérée comme illégale et le coup ne peut pas être joué.

Dès qu'un joueur a capturé 25 graines ou plus, il est déclaré vainqueur de la partie et le jeu s'arrête.

Dans le cas où un joueur ne peut pas jouer, la partie s'arrête et toutes les graines restant sur le tablier sont capturées par son adversaire.  Dans ce cas, le joueur qui a capturé le plus de graines a gagné.

### Représentation du tablier

Le tablier est toujours représenté avec les cases de Sud en bas.  
Les cases sont ordonnées et repérées par des chiffres, de 1 à 6, de la gauche vers la droite. Dans chaque case, on indique le nombre de graines qu'elle contient. Ainsi, le tablier de départ classique est donné ci-dessous:

Un coup est noté en donnant le camp qui joue ainsi que le numéro de la case qui est vidée.  Par exemple, $($SUD$, 2)$ si le camp Sud joue en prenant les graines de la deuxième case en partant de la gauche qui se trouve dans son camp. Ou alors, $($NORD$, 6)$ si le camp Nord joue en prenant les graines de la case numéro 6 qui se trouve dans son camp, c'est à dire la sixième case en partant de la gauche qui se trouve dans son camp, soit celle la plus en haut et à droite du tablier.
