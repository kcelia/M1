```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import math
```

```python
def make_grid(xmin=-5,xmax=5,ymin=-5,ymax=5,step=20,data=None):
    """ Cree une grille sous forme de matrice 2d de la liste des points
    :return: une matrice 2d contenant les points de la grille, la liste x, la liste y
    """
    if data is not None:
        xmax,xmin,ymax,ymin = np.max(data[:,0]),np.min(data[:,0]),\
                              np.max(data[:,1]),np.min(data[:,1])
    x,y = np.meshgrid(np.arange(xmin,xmax,(xmax-xmin)*1./step),
                      np.arange(ymin,ymax,(ymax-ymin)*1./step))
    grid=np.c_[x.ravel(),y.ravel()]
    return grid, x, y

def load_usps(filename):
    with open(filename,"r") as f:
        f.readline()
        data =[ [float(x) for x in l.split()] for l in f if len(l.split())>2]
    tmp = np.array(data)
    return tmp[:,1:],tmp[:,0].astype(int)
```

```python
def optimize(fonc, dfonc, xinit, eps, max_iter):
    """fonc : la fonction a optimiser
       dfonc : le gradient de la fonction à optimiser
       xinit : point initial
       eps : le pat du gradient
       max_iter : le nombre d'itération maximum
       retourne la liste des points au temps t, f(x), et le gradient de cette fonction"""
       
    x = []
    x.append(xinit)
    f = []
    df = []
   
    for t in range(max_iter):
        tmp= []
        tmpd = []
        for i in range(len(xinit)):
            tmp.append(fonc(x[t][i]))
            tmpd.append(dfonc(x[t][i]))
        if(t!=max_iter-1):
            x.append(np.array(x[t]) - eps*np.array(tmpd))
        f.append(tmp)
        df.append(tmpd)
    return np.array(x), np.array(f), np.array(df)
```

```python
#fonction 1
def function1(x):
    """float -> float
       retourne x*cos(x)"""
    return x*math.cos(x)
    
def dfunction1(x):
    """float->float
       retourne la dérivée de x pour la fonction 1"""
    return -math.sin(x)*x + math.cos(x)
    
#fonction 2
def function2(x):
    """float->float
       retourne -log(x)+x*x"""
    return -math.log(x)+x**2
    
def dfunction2(x):
    """float->float
       retourn la dérivée de x pour la fonction 2"""
    return -1.0/x + 2*x
    
#fonction 3
def function3(x):
    """float->float
       return 100*(x2-x1**2)**2 + (1-x1)**2"""
    return 100*(x[1] - x[0]**2)**2 + (1 - x[0])**2
    
def dfunction3(x):
    """float->float
       retourne la dérivée par rapport à x1 de la fonction 3 et par rapport à x2"""
       
    dx1 =-400*x[0]*(x[1]-x[0]**2)-2*(1-x[0])# x[0]*(-400*x[1] + 2 + 400*(x[0]**2))
    dx2 = 200*(x[1]-(x[0]**2))
    return [dx1,dx2]
```

```python
#fonction 1
nb_it = 300
x, f, df = optimize(function1, dfunction1, [6], 0.01, nb_it)
plt.title("fonction 1")
plt.plot(range(nb_it), x, label="x")
plt.plot(range(nb_it), f, label="f")
plt.plot(range(nb_it), df, label="df")
l = np.log(np.absolute(x-x[nb_it-1]))
plt.plot(range(nb_it), l, label="log(||x-x[nb_it]||)")
plt.legend()
plt.show()
```

```python
def optimize_log(fonc, dfonc, xinit, eps, max_iter, datax):
    """fonc : la fonction a optimiser
       dfonc : le gradient de la fonction à optimiser
       xinit : point initial w0
       eps : le pat du gradient
       datax : base de données
       max_iter : le nombre d'itération maximum
       retourne la liste des points au temps t, f(x), et le gradient de cette fonction"""
       
    x = []
    x.append(xinit)
    f = []
    df = []
   
    for t in range(max_iter):
        f.append(fonc(x[t], datax[t]))
        df.append(dfonc(x[t], datax[t]))
        if(t!=max_iter-1):
            x.append(np.array(x[t]) - eps*np.array(df[t]))
    return np.array(x), np.array(f), np.array(df)
```

```python
def fw(w,x):
    """HYP : w doit avoir un rang de plus que x
       w : vecteur des paramètre
       x : vecteurs des données"""
    return w[0] + np.vdot(w[1:],x)

def dfw(w, x):
    dw = []
    dw.append(1)
    for i in range(1,len(w)):
        dw.append(x[i-1])
    return dw
```

```python
class regressionLogistique:
    def __init__(self, nb_it, grad):
        self.nb_it = nb_it
        self.grad = grad
        
    def fit(self, datax):
        w0 = [1 for i in range(len(datax[0])+1)]
        x, f, df = optimize_log(fw, dfw, w0, self.grad, self.nb_it, datax)
        self.w_et = x[len(x)-1]
        #print("self.w_et", self.w_et)
    
    def predict(self, datax):
        fw = []
        tmpt = []
        for i in range(len(datax)):
            tmp = self.w_et[0] + np.vdot(self.w_et[1:], datax[i])
            mu = 1.0/(1+ math.exp(-tmp))
            print("mu",mu)
            tmpt.append(tmp)
            if(mu < 1):
                fw.append(-1)
            else:
                fw.append(1)
        self.pred = fw
        #print("self.pred",self.pred)
        print("tmpt",max(tmpt))
        
    def accuracy(self, datay):
        scor = 0
        #print("datay", datay)
        for i in range(len(datay)):
            if(self.pred[i] == datay[i]):
                scor += 1
        return scor*1.0/len(datay)
```

```python
def classe_binaire(datax, datay, classe1, classe2):
    """[float*float]*[float]*int*int->[float*float]*[float]
       HYP : classe1 et classe2 sont compris entre 0 et 9 compris
       Cette fonction renvoie un nouveau datax et datay ne sélectionnant que les exemples ayant que pour classe classe1 ou
       classe2."""
    
    new_datax = []
    new_datay = []
    for i in range(len(datax)):
        if(datay[i]==classe1): 
            new_datax.append(datax[i])
            new_datay.append(-1)
        elif(datay[i]==classe2):
            new_datax.append(datax[i])
            new_datay.append(1)
    return new_datax, new_datay
```

```python
dataxtrain, dataytrain = classe_binaire(dataxtrain, dataytrain, 2, 5)
dataxtest, dataytest = classe_binaire(dataxtest, dataytest, 2, 5)
```
